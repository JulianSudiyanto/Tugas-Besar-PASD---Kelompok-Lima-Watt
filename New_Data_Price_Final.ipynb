```python
import pandas as pd
```

```python
df = pd.read_csv("restaurant_sales_data.csv")
df.head(200)
```

```python
print("Jumlah data awal:", len(df))
```

```python
df = df.drop(columns=['Order Date'], errors='ignore')
```

```python
duplicate_count = df.duplicated().sum()
print("Jumlah duplikasi:", duplicate_count)
```

```python
df = df.drop_duplicates()
```

```python
print("Jumlah nilai kosong per kolom:")
print(df.isnull().sum())
```

```python
df = df.dropna()
```

```python
print("Jumlah nilai kosong per kolom:")
print(df.isnull().sum())
```

```python
df_200 = df.head(200)
```

```python
df_unique = df_200.drop_duplicates()
```

```python
df_unique.to_csv("Cleaned_200_Price_Data_Final.csv", index=False)
```

```python
print("Jumlah data setelah dibersihkan dan dipotong 200:", len(df_unique))
```

```python
df_200 = df.head(200).drop_duplicates().copy()

conversion_rate = 15800
df_200['Price'] = df_200['Price'] * conversion_rate
df_200['Price'] = df_200['Price'].astype(int)

df_200['Order Total'] = df_200['Price'] * df_200['Quantity']
df_200['Order Total'] = df_200['Order Total'].astype(int)

df_200.to_csv("Cleaned_200_Price_Data_Final.csv", index=False)
```
